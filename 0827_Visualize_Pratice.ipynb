```python
X = np.arange(1, 10)
y1 = X
y2 = X**2
y3 = X ** 3 - X
label = ['y1', 'y2','y3']
```

```python
plt.figure(figsize = (8,4))
plt.plot(X, y1,'r--', X, y2, 'bs',  X, y3,'g^',) # 서로 다른 마커를 넣기 위해서 사용
# plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.legend(label, loc = 'best') # 각각을 라벨링 하기 위해서, 갯수 맞아야 할ㄷ스
```

```python
# 박스 플롯을 쉽게 구현하기 위해 다른 데이터 셋 활용
tips = sns.load_dataset('tips')
```

```python
print (tips.shape)
print (tips.info())
print (tips.head())
```

```python
sns.boxplot(tips['tip'])
```

```python
tips.columns
```

```python
sns.boxplot(tips['time'], tips['total_bill'])
# 시간에 따른 총 비용이 얼마인지 알아보자
```

```python
sns.boxplot(tips['time'], tips['total_bill'], hue = tips['sex'])
# data = 파라미터를 입력 안했기 때문에 df[] 이런식으로 입력하였음.
# data 파라미터 입력시 아래처럼 하면 됨.
```

```python
 ax = sns.boxplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set3")
```

```python
 ax = sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
#palette 파라미터는 색을 결정하는 듯.
```

```python
sns.boxplot(x = 'time', y ='tip', hue = 'sex', data = tips, order = ['Dinner', 'Lunch'])
# order 파라미터는 순서를 결정할 수 있다.
```

```python
sns.boxplot(x = 'day', y = 'tip', hue = 'smoker', data = tips, order = ['Sun', 'Fri'])
# order를 전체가 아닌 일부를 입력할 경우 그 해당 입력분만 출력된다.
```

```python
sns.swarmplot(x = 'day', y = 'tip', data= tips)
# boxplot 형태의 scatter 버젼
```

```python
sns.swarmplot(x = 'sex', y = 'total_bill', hue = 'smoker', data = tips)
```

```python
plt.figure(figsize = (10, 4))
# 안해줘도 되나 그래프의 사이즈를 직접 선택하고 싶을 경우 하는게 좋다.
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
# 두가지가 섞였으나 보기가 너무 안좋다. 색을 바꿔보고싶다.
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'black', edgecolor = 'red')
# color 파라미터를 통해서 색을 바꿔보았다.
# edgecolor 파라미터가 있는데 이건 또 뭘까 : 테두리 색깔인듯
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
# boxplot과는 다른 형태
```

```python
# lmplot를 사용
sns.lmplot(x = 'tip', y = 'total_bill', data = tips)
```

```python
sns.lmplot(x = 'tip', y = 'total_bill', hue = 'sex', data = tips)
# line 위 아래의 음영이 의미하는걸 모르겠다.
# X,y값은 당연히 int 여야한다.
```

```python
sns.lmplot(x = 'tip', y = 'total_bill', hue = 'day', data = tips)
```

```python
sns.lmplot(x = 'tip', y = 'total_bill', hue = 'sex',  scatter_kws={'marker':'D', 's': 100}, data = tips)
# scatter_kws 는 딕트형태로 입력되며 s는 size / marker 도 가능
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue = 'species', vars = ['sepal_length', 'sepal_width', 'petal_length'] )
# columns이 추가 되면서 hue에 입력이 가능 해졌다.
# vars 는 리스트로 parameter를 받고 해당 부분만을 뽑아내준다.
```

```python
# jointplot : tips
sns.jointplot(x = 'tip', y = 'total_bill', data = tips)
# scatter와 hist까지 같이 보여줌
# kind 로 모양을 설정할수 있는데 'reg', 'hex', 'kde' 등이 있다
```
