```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
# sklearn으로 부터 불러온다
# 클래스로부터 객체를 생성한다.
# 매개변수를 입력한다(K의 갯수)
```

```python
knn.fit(X_train, y_train) 
# 함수를 훈련 시키기 위해 fit 한다
# train 함수의 x 값과 y 값
#
```

```python
X_new = np.array([[5,2.9,1,0.2]])
```

```python
prediction = knn.predict(X_new)
print ('predict : {}'.format(prediction))
print (iris_dataset['target_names'][prediction])
# 예측값은 0,1,2, 로 구분지어서 나온다.
```

```python
y_pred = knn.predict(X_test)
# X테스트 값을 넣어서 0,1,2, 의 y 값을 뽑아 낸다.
```

```python
print ('정ㅇ확도 : {}'.format(np.mean(y_pred == y_test)))
# 실제 값과 예측 값의 맞는 갯수의 평균.
```

```python
knn.score(X_test, y_test)
# x_test는 곧 x 값에 대한 y의 예측 값을 뜻함
# 즉 y_pred와 y_test 를 비교하고 있다.
```
