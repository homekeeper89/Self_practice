```python
from numpy.random import randn
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(randn(50).cumsum(), 'k--')
# k-- 옵션은 검은 점선을 나타낸다.
# fig.add_subplot의 반환 객체는 ax 객체이며 각각을 직접 호출 가능하다
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(randn(50).cumsum(), 'k--')
_ = ax1.hist(randn(100), bins = 20, color = 'k', alpha  = 0.10)
# alpha 는 그래프의 투명도(진한정도)를 나타낸다. 0-1사이의 값으로 정할 수 있다.
ax2.scatter(np.arange(30), np.arange(30)+3 *randn(30), alpha = 0.1)
# scatter 역시 alpha 파라미터가 존재함
```

```python
fig, axes = plt.subplots(2,3, figsize = (8, 6))
# 2행 3열의 형태의 그래프 자리는 만든다
# figsize를 통해 그래프 크기를 정할 수 있다.
# 각 그래프를 어떻게 불러오지? axex[1,1] 이런식으로 가능한듯. 단 0-1, 0-2 까지 인듯함.(2,3 사이즈 이므로)
```

```python
# 서브플롯 간 간격 조절하기
fig, axes = plt.subplots(2,2, sharex = True, sharey = True)
# sharex, sharey : 모든 서브플롯이 같은 x, y축 눈금을 사용하도록 한다.
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins = 50, color = 'k', alpha = 0.8)
plt.subplots_adjust(wspace = 0, hspace = 0) # 서브플롯간 간격을 조절하는 메쏘드
# 4개의 서브플롯에 랜덤의 500개 값이 들어감.
```

## 눈금, 라벨, 범례

```python
# xlim, xticks, xticklabels 등의 메써드가 있다.
```

```python
fig = plt.figure(); ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
```

```python
# X축 눈금을 변경하기 위한 가장 쉬운 방법은 set_xticks, set_xticklabels 이다
# set_xticks는 ㅈㄴ체 데이터 범위에 맞춰 눈금을 어디에 배치할지 지정한다. 
# 차후 이 위치에 눈금 이름이 들어간다.
# 다른 눈금 이름을 지정하고 싶다면 set_xticklabels를 사용하면 된다
```

```python
fig = plt.figure(); ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation = 30, fontsize = 'small')
```

## 범례 추가하기

```python
# 각 그래프별 label 인자를 남기자

fig = plt.figure(); ax = fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'k', label = 'one')
ax.plot(randn(1000).cumsum(), 'k--', label = 'two')
ax.plot(randn(1000).cumsum(), 'k.', label = 'three', alpha = 0.3)
# alpha는 여기서도 통함
```

# Pandas에서 그리기

```python
s = pd.Series(np.random.randn(10).cumsum(), index = np.arange(0,100, 10))
s.plot()
# plot() 메써드를 통해서 단순 선 그래프를 그릴 수 있다.
# series에선 index가 기본적으로 x축으로 인식됨
```

```python
df = pd.DataFrame(np.random.randn(10,4).cumsum(0), columns = ['a', 'b', 'c', 'd'], index = np.arange(0,100,10))
```

```python
df.plot(grid = True)
# 각 칼럼별로 그래프가 한번에 그려져셔 나왔다.
# plot()에 숫자로 인자가 들어가지는데 무슨 의미인지 감이 안잡힘
```

```python
df.plot(3, grid = True)
# 각 숫자가 칼럼을 의미함. 그 해당 칼럼을 x축으로 옮기는것 같은데 그래프가 이해는 안감
```

## 막대 그래프 그리기

```python
# kind 파라미터를 활용하자. bar / barh
fig, axes = plt.subplots(2,1, figsize = (8,6))
data = pd.Series(np.random.rand(16), index = list('abcdefghijklmnop'))
data.plot(kind = 'bar', ax = axes[0], color = 'k', alpha = 0.7)
data.plot(kind = 'barh', ax = axes[1], color = 'k', alpha = 0.7)
```

```python
df = pd.DataFrame(np.random.randn(6,4), index = np.arange(6), columns = pd.Index(['a', 'b', 'c', 'd']))
# np.로 만들경우 인덱스*칼럼 이라는 것을 기억하자.
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
```

```python
party_counts
```

```python
party_counts_df = party_counts.loc[:, 2:5]
# 혼자거나 6명인 파티는 제외하는 목적
```

```python
party_counts_df
```

```python
party_pacs = party_counts_df.div(party_counts_df.sum(1).astype(float), axis =0)
# div는 뭘까.
```

```python
party_pacs.plot(kind = 'bar', stacked = True)
```
