```python
plt.plot(data, drawstyle = 'steps-post')
```

```python
fig = plt.figure(figsize = (10,5))
plt.plot(data, drawstyle = 'steps-post', label = 'step')
plt.plot(data, 'k', label = 'default')
plt.legend(loc = 'best')
```

```python
fig = plt.figure(figsize = (10, 8))
plt.plot(np.random.rand(30), 'k', label = 'k, default')
plt.plot(np.random.rand(30), 'ko', label = 'k0')
plt.plot(np.random.rand(30), 'b--', label = 'k--')
plt.plot(np.random.rand(30), 'bo--', label = 'k0--')
plt.legend(loc = 'best')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(data, 'ko')
ax.set_xticks([0,5,10])
ax.set_xticklabels(['one', 'two', 'three'], rotation = 30)
```
